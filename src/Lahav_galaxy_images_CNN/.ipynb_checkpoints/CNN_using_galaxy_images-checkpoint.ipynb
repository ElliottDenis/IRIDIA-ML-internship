```python
# basics
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# creating train and test sets
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# neural network
import tensorflow as tf
from tensorflow import keras
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
```

```python
# CONSTANTS

# number galaxies to select out of the total dataset
N_GALAXIES = 10000

# for training (batch size = number of points to average before back propagation)
BATCH_SIZE = 256
EPOCHS = 20

# specifying optimiser we will be using (could try SGD)
LEARNING_RATE = 0.01
OPT = tf.keras.optimizers.Adam(learning_rate = LEARNING_RATE)
```

```python
def get_train_test():
    
    '''GET DATASET FROM CSV FILE, REMOVE OUTLIERS AND DROP USELESS COLUMNS,
    ONLY KEEP FILTERS AND REDSHIFTS. ONLY SELECT FIRST 10,000 GALAXIES'''
    
    # load
    df = pd.read_csv("../../data/photo_z/df_for_32x32_images.csv")
    
    # select first 10,000 galaxies
    df = df.iloc[range(N_GALAXIES)]
    
    # remove outliers (we noticed some objects had values for u/z/g/redshift below zero)
    outliers = ((df["petroMag_u"] <= 0)  
                | (df["petroMag_z"] <= 0) 
                | (df["petroMag_g"] <= 0)
                | (df["petroMag_r"] <= 0)
                | (df["petroMag_i"] <= 0))

    # drop the row corresponding to the outlier
    df = df[~outliers]
    
    # we drop the data we are not interested in and which won't be useful
    # expect isotropic universe -> also drop position in sky
    df = df.drop(
        columns=["old_index", "objID", "run", "rerun", "camcol", "field", "rowc", 
                 "colc", "fieldID", "image_path", "csv_path_corrected", "objID_corrected",
                "image_path_corrected"], axis=1)
    
    
    # replace and drop all NaN
    df = df.replace([np.inf, -np.inf], np.nan)
    df = df.dropna(axis=0)
    df = df.reset_index(drop = True)

    # array of galaxy images
    arr = np.load("../../data/photo_z/galaxy_images_32x32_all.npy")
    arr = arr[range(N_GALAXIES)]
    arr = arr[~outliers]
    
    return df, arr

# get data
df, arr = get_train_test()
```

```python
# seperate between input (photometric values) and outputs (redshifts)
x = df.drop(columns = "zspec")
y = df["zspec"]
```

```python
# easy train-test split (with photometric data AND images)
x_train, x_test, arr_train, arr_test, y_train, y_test = train_test_split(x, arr, y)
```

Images in input - 3 epochs that trains -> regression of redshift (will try adding filters later)

input_shape = (32,32,5)

one output layer -> regression (can be negative!!) -> activation function linear

con-pull-conv-pull -> flatten -> dense(ReLu)-dense(ReLu)-dense(linear)

```python
def plot_loss_acc(loss, val_loss):

    plt.plot(loss, label="MSE training set")
    plt.plot(val_loss, label="MSE validation set")
    plt.grid(True)
    plt.title("mean_squared_error")
    plt.legend()
    
    return
```

```python
# if we just use the images without adding the photometric measurements
model_CNN = keras.models.Sequential([
    keras.layers.Conv2D(5,2, strides=2, activation="relu", input_shape=[32,32,5]),
    keras.layers.AvgPool2D(2),
    keras.layers.Conv2D(10,2, activation="relu"),
    keras.layers.MaxPool2D(2),
    keras.layers.Flatten(),
    keras.layers.Dense(5, activation="relu"),
    keras.layers.Dense(5, activation="relu"),
    keras.layers.Dense(1, activation="linear")
])
```

```python
# multi-input neural network, adding photometric data after flatten of images
input_img = keras.Input(shape=(32, 32, 5))
x = keras.layers.Conv2D(5,2, strides=2, activation="relu", input_shape=[32,32,5])(input_img)
x = keras.layers.MaxPool2D(2)(x)
x = keras.layers.Conv2D(10,2, activation="relu")(x)
x = keras.layers.MaxPool2D(2)(x)
x = keras.layers.Flatten()(x)
input_photo = keras.Input(shape=(5))
x = keras.layers.Concatenate()([x, input_photo])
x = keras.layers.Dense(128, activation="relu")(x)
x = keras.layers.Dense(64, activation="relu")(x)
x = keras.layers.Dense(1, activation="linear")(x)

model = keras.Model(inputs = [input_img, input_photo], outputs = x)
```

```python
model.compile(optimizer = OPT, loss='mean_squared_error',metrics=['mean_squared_error'])
model.summary()
```

```python
history = model.fit([arr_train, x_train], y_train, 
                           epochs = EPOCHS, 
                           batch_size = BATCH_SIZE,
                           validation_data=([arr_test, x_test], y_test)
                          )
```

```python
hist_loss = history.history["mean_squared_error"]
hist_val_loss = history.history["val_mean_squared_error"]

plot_loss_acc(hist_loss, hist_val_loss)
```
